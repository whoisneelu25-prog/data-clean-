# 🧹 Data Cleaning Walkthrough
This notebook walks through every cleaning step on the raw employee dataset.

**Issues we fix:**
- Duplicate rows
- Inconsistent text casing
- Invalid emails
- Mixed date formats
- Out-of-range numeric values
- Missing values
- Malformed phone numbers

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../scripts')
from data_cleaning import *

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
```

## 1. Load Raw Data

```python
df = pd.read_csv('../data/raw/employees_raw.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Initial Audit

```python
issues = audit(df)
print('Missing values per column:')
print(pd.Series(issues['missing_per_col']))
print(f"\nDuplicate rows: {issues['duplicate_rows']}")
```

## 3. Run Full Cleaning Pipeline

```python
df_clean = clean(df.copy())
```

## 4. Review Cleaned Data

```python
print(f'Shape after cleaning: {df_clean.shape}')
df_clean
```

## 5. Compare Missing Values Before vs After

```python
comparison = pd.DataFrame({
    'Before': df.isnull().sum(),
    'After':  df_clean.isnull().sum()
}).fillna(0).astype(int)
comparison['Resolved'] = comparison['Before'] - comparison['After']
comparison
```

## 6. Export Cleaned Data

```python
df_clean.to_csv('../data/cleaned/employees_cleaned.csv', index=False)
print('Saved → data/cleaned/employees_cleaned.csv')
```
